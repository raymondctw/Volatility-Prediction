```python
import Data
import pandas as pd
import numpy as np
import talib
```

```python
historical_data = Data.get_historical_data('QQQ').dropna()
historical_data['Return'] = historical_data['Close'].pct_change().dropna()
historical_data['absReturn'] = historical_data['Return'].abs() * np.sqrt(252)
historical_data['Volatility(M)'] = talib.EMA(historical_data['absReturn'], timeperiod=22).dropna()
```

```python
# Visualize the daily vol and EMA vol
historical_data['absReturn'].plot(kind='area',figsize=(20, 10))
historical_data['Volatility(M)'].plot(kind='line')
```
